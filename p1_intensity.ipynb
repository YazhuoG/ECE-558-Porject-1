```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import time
```

```python
# use OpenCV to read dimensional values of image
img = cv2.imread("wolves.png", cv2.IMREAD_GRAYSCALE)
# cv2.imshow("image", img)
# cv2.waitKey(0)
print('Shape: ', img.shape)
print('Size: ', img.size)
```

```python
# size of image matrix
y_axis = len(img)
x_axis = len(img[0])
Intensity_diff1 = []
Intensity_diff2 = []
start = np.zeros(4)
end = np.zeros(4)
```

```python
start[0] = time.time()
for i in range(y_axis-1):
    for j in range(x_axis):
        Intensity_diff1.append((img[i][j] - img[i+1][j])**2)
end[0] = time.time()
```

```python
start[1] = time.time()
for i in range(y_axis):
    for j in range(x_axis-1):
        Intensity_diff2.append((img[i][j] - img[i][j+1])**2)
end[1] = time.time()
```

```python
start[2] = time.time()
plt.hist(Intensity_diff1, 50, label="(x,y+1)")
plt.title("Intensity pair (x,y) & (x,y+1)")
end[2] = time.time()
```

```python
start[3] = time.time()
plt.hist(Intensity_diff2, 50, label="(x+1,y)")
plt.title("Intensity pair (x,y) & (x+1,y)")
end[3] = time.time()
```

```python
tot_time = sum(end) - sum(start)
print("Total runtime is", tot_time, "seconds.")
```
