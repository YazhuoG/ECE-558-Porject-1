```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import time
```

```python
# use OpenCV to read dimensional values of image
img = cv2.imread("wolves.png", cv2.IMREAD_UNCHANGED)
img = img[:, :, ::-1] # BGR to RGB
# imgplot = cv2.imshow("image", img)
# cv2.waitKey(0)
print('Shape: ', img.shape)
print('Size: ', img.size)
```

```python
# size of image matrix
y_axis = len(img)
x_axis = len(img[0])
z_axis = len(img[0][0])
BGR_diff1 = []
BGR_diff2 = []
BGR_diff3 = []
start = np.zeros(6)
end = np.zeros(6)
```

```python
start[0] = time.time()
for i in range(y_axis-1):
    for j in range(x_axis):
        sum_diff = 0
        for k in range(z_axis):
            sum_diff += (img[i][j][k] - img[i+1][j][k])**2
        BGR_diff1.append(sum_diff)
end[0] = time.time()
```

```python
start[1] = time.time()
for i in range(y_axis):
    for j in range(x_axis-1):
        sum_diff = 0
        for k in range(z_axis):
            sum_diff += (img[i][j][k] - img[i][j+1][k])**2
        BGR_diff2.append(sum_diff)
end[1] = time.time()
```

```python
start[2] = time.time()
for i in range(y_axis-1):
    for j in range(x_axis-1):
        sum_diff = 0
        for k in range(z_axis):
            sum_diff += (img[i][j][k] - img[i+1][j+1][k])**2
        BGR_diff3.append(sum_diff)
end[2] = time.time()
```

```python
start[3] = time.time()
plt.hist(BGR_diff1, 50, label="(x,y+1)")
plt.title("RGB pair (x,y) & (x,y+1)")
end[3] = time.time()
```

```python
start[4] = time.time()
plt.hist(BGR_diff2, 50, label="(x+1,y)")
plt.title("RGB pair (x,y) & (x+1,y)")
end[4] = time.time()
```

```python
start[5] = time.time()
plt.hist(BGR_diff3, 50, label="(x+1,y+1)")
plt.title("RBG pair (x,y) & (x+1,y+1)")
end[5] = time.time()
```

```python
tot_time = sum(end) - sum(start)
print("Total runtime is", tot_time, "seconds.")
```
